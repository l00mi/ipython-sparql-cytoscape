# SPARQL to GRAPH

```python
import sparql_dataframe
import pandas as pd
from ipycytoscape import *
```

```python
endpoint = "https://lindas.admin.ch/query"

q = """
    SELECT *
    FROM <https://lindas.admin.ch/meta>
    WHERE {
        ?s ?p ?o
    }
    limit 1000
"""

df = sparql_dataframe.get(endpoint, q)
```

```python
graph_nodes = []
graph_edges = []

for x in df.values:
    graph_nodes.append(Node(data={'id': x[0] , 'label': x[0]}))
    graph_nodes.append(Node(data={'id': x[2] , 'label': x[2]}))
    graph_edges.append(Edge(data={'id': x[0]+x[1]+x[2], 'label': x[1], 'source': x[0], 'target': x[2]}))
```

```python
cytoscapeobj = CytoscapeWidget()
cytoscapeobj.set_layout(name='cola')

cytoscapeobj.set_style([                        {
                        'selector': 'node',
                        'style': {
                                  'font-family': 'helvetica',
                                  'font-size': '20px',
                                  'label': 'data(label)',
                                  'text-valign': 'bottom',
                                  'text-halign': 'center'                            
                        }   
                      },
                    {
                        'selector': 'edge',
                        'style': {
                                  'font-family': 'helvetica',
                                  'font-size': '20px',
                                  'label': 'data(label)',
                                  'text-valign': 'bottom',
                                  'text-halign': 'right'                            
                        } 
                    }
])

cytoscapeobj.graph.add_nodes(graph_nodes)
cytoscapeobj.graph.add_edges(graph_edges)
cytoscapeobj
```
